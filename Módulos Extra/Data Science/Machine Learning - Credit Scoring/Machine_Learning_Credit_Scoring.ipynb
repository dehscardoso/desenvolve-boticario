### RegressÃ£o LogÃ­stica

```python
y = dados['default']
x = dados.drop(columns = 'default')
```

```python
from sklearn.model_selection import train_test_split

SEED = 77
treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size = 0.30, random_state = SEED)
```

```python
base_treino = treino_x.shape[0]
base_teste = teste_x.shape[0]

print('A base de teste de treino tem %s elementos e a base de teste tem %s elementos.' %(base_treino, base_teste))
```

```python
from sklearn.linear_model import LogisticRegression

modelo = LogisticRegression(max_iter=1000)
modelo.fit(treino_x, treino_y)
print('AcurÃ¡cia de treino: ', modelo.score(treino_x, treino_y))
```

```python
previsoes = modelo.predict(teste_x)
previsoes
```

```python
from sklearn.metrics import accuracy_score

acuracia = accuracy_score(teste_y, previsoes)
acuracia = round(acuracia, 3)*100
print('AcurÃ¡cia de teste: ', acuracia)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix

plot_confusion_matrix(modelo, teste_x, teste_y, cmap='Blues', values_format = '.3g');
```

---
### PrecisÃ£o
ProporÃ§Ã£o de prediÃ§Ãµes corretas de uma categoria em relaÃ§Ã£o a todas as prediÃ§Ãµes desta categoria.

**PrecisÃ£o = VP / VP + FP**

---
### Recall
ProporÃ§Ã£o de prediÃ§Ãµes corretas de categoria alvo.

**Recall = VP / VP + FN**

----
### F1-Score
MÃ©dia harmÃ´nica entre precisÃ£o e recall.

**F1 - score = 2 * (PrecisÃ£o * Recall) / (PrecisÃ£o + Recall)**

----

```python
from sklearn.metrics import classification_report

print(classification_report(teste_y, previsoes))
```

---
### Curva ROC

Avalia o desempenho do modelo a partir da relaÃ§Ã£o entre a taxa de verdadeiros positivos (sensibilidade) e a taxa de falsos positivos (especificidade) para diferentes limites.

**Especificidade = VN / (VN + FP)**

**Sensibilidade = VP (VP + FN)**

---

```python
from sklearn.metrics import roc_curve, roc_auc_score

prob_previsao = modelo.predict_proba(teste_x)[:,1]

# tfp = taxa falsos positivos
# tvp = taxa verdadeiros positivos

tfp, tvp, limite = roc_curve(teste_y, prob_previsao)
print('roc_auc', roc_auc_score(teste_y, prob_previsao))

plt.subplots(1, figsize=(5,5))
plt.title('Curva ROC')
plt.plot(tfp,tvp)
plt.xlabel('Especifidade')
plt.ylabel('Sensibilidade')
plt.plot([0, 1], ls="--", c = 'red') # plotando linha guia pontilhada vermelha
plt.plot([0, 0], [1, 0], ls="--", c = 'green'), plt.plot([1, 1], ls="--", c = 'green'); # plotando linha guia pontilhada verde

# Ã¡rea > 0.8, entÃ£o discriminaÃ§Ã£o excelente!
```

---
### AUC

Quantifica a Curva ROC atravÃ©s da Ã¡rea sob esta curva

ConvenÃ§Ã£o EstatÃ­stica:

- Ã¡rea < 0,7: baixa discriminiaÃ§Ã£o
- 0,7 â‰¤ Ã¡rea < 0,8: discriminaÃ§Ã£o aceitÃ¡vel
- 0,8 â‰¤ Ã¡rea < 0,9: discriminaÃ§Ã£o excelente
- Ã¡rea > 0,9: discriminaÃ§Ã£o excepcional

```python
# categorias da variavel 'sexo_est_civil'

# male   : divorced/separated = 1
# female : divorced/separated/married = 2
# male   : single = 3
# male   : married/widowed = 4
# female : single = 5

# criando a variÃ¡vel 'sexo'

## 1 = mulher
## 0 = homem

# criando a variÃ¡vel 'estado_civil'
## 1 = solteira
## 0 = casada/divorciada/viÃºva

dic_sexo = {2:1, 5:1, 1:0, 3:0, 4:0}
dic_est_civil = {3:1, 5:1, 1:0, 2:0, 4:0}

dados['sexo'] = dados['sexo_est_civil'].map(dic_sexo)
dados['estado_civil'] = dados['sexo_est_civil'].map(dic_est_civil)

dados.head()
```

```python
# funÃ§Ã£o para retirar colunas da base de dados

colunas_retiradas = ['sexo_est_civil', 'sexo']
base_dados = dados.drop(columns=colunas_retiradas)
base_dados.head()
```

```python
# funÃ§Ã£o 'roda_modelo': ajuste a validaÃ§Ã£o

def roda_modelo(base_dados, classificador):

# separando a base de modelagem e variÃ¡vel resposta

    y = base_dados['default']
    x = base_dados.drop(columns='default')

# separando a base de treino e teste

    SEED = 77
    treino_x, teste_x, treino_y, teste_y = train_test_split(x, y, test_size=0.30, random_state = SEED)

    base_treino = treino_x.shape[0]
    base_teste = teste_x.shape[0]
    print('A base de treino tem %s elementos e a base de teste tem %s elementos.' %(base_treino, base_teste))
    print(100*'-')

# ajustamdo modelo com base de teste
    
    modelo = classificador
    modelo.fit(treino_x, treino_y)

#-------------------------------------------------------------------------------------------------------------------------------------

# matriz de confusÃ£o

    matriz_confusao = plot_confusion_matrix(modelo, teste_x, teste_y, values_format = '.3g', cmap = 'Blues')
    plt.title('Matriz de Confusao')
    plt.show

# classification report

    previsoes = modelo.predict(teste_x)

    print('\nClassification Report:')
    print(classification_report(teste_y, previsoes))

# curva ROC  e AUC

    print(100*'-')
    prob_previsao = modelo.predict_proba(teste_x)[:,1]
   
    tfp, tvp, limite = roc_curve(teste_y, prob_previsao)
    print('roc_auc:', roc_auc_score(teste_y, prob_previsao))
    
    plt.subplots(1, figsize=(5,5))
    plt.title('Curva ROC')
    plt.plot(tfp, tvp)
    plt.plot([0, 1], ls="--", c = 'red') # plotando linha pontilhada guia para regressao aleatÃ³ria
    plt.plot([0, 0], [1, 0], ls="--", c = 'green'), plt.plot([1, 1], ls="--", c = 'green') # plotando linha pontilhada guia para regressÃ£o perfeita
    plt.ylabel('Sensibilidade')
    plt.xlabel('Especificidade')
    plt.show() 

    return modelo, matriz_confusao
```

```python
roda_modelo(base_dados, LogisticRegression(max_iter=1000))
```
